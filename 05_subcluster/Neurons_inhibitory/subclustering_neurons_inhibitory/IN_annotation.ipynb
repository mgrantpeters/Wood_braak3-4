# 1.0 First checks

```python
import muon as mu
import numpy as np
import scanpy as sc
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mudata = mu.read('Neurons_Inhibitory_clustered.h5mu')
```

```python
mudata['rna'].obsm['X_umap'] = mudata.obsm['X_umap']
mudata['rna'].obs['mm_leiden_res0.7'] = mudata.obs['leiden_res0.7']
```

```python
genes = ['mm_leiden_res0.7', 'Group',  'phase', 'G2M_score', 'region','sample_id']

sc.pl.umap(mudata['rna'], color = genes)
```

```python
tmp = pd.crosstab(mudata['rna'].obs['mm_leiden_res0.7'],mudata['rna'].obs['sample_id'],normalize='index')
tmp.plot.bar(stacked=True).legend(bbox_to_anchor=(1.0, 1.0))
plt.show()

tmp = pd.crosstab(mudata['rna'].obs['mm_leiden_res0.7'],mudata['rna'].obs['Group'],normalize='index')
tmp.plot.bar(stacked=True).legend(bbox_to_anchor=(1.0, 1.0))
plt.show()
```

# 4.0 Annotation transfer and final checks

```python
mudata['rna'].obs['annotation_level_0'] = mudata['rna'].obs['mm_leiden_res0.7'].replace({
    '0' : 'Neurons',
    '1' : 'Neurons',
    '2' : 'Neurons',
    '3' : 'Neurons',
    '4' : 'Neurons',
    '5' : 'Neurons',
    '6' : 'Neurons',
    '7' : 'Neurons',
    '8' : 'Neurons',
    '9' : 'Neurons',
    '10' : 'Neurons',
    '11' : 'Neurons',
    '12' : 'Neurons'
})

mudata['rna'].obs['annotation_level_1'] = mudata['rna'].obs['mm_leiden_res0.7'].replace({
    '0' : 'IN',
    '1' : 'IN',
    '2' : 'IN',
    '3' : 'IN',
    '4' : 'IN',
    '5' : 'IN',
    '6' : 'IN',
    '7' : 'IN',
    '8' : 'IN',
    '9' : 'IN',
    '10' : 'IN',
    '11' : 'IN',
    '12' : 'IN'
})

mudata['rna'].obs['annotation_level_2'] = mudata['rna'].obs['mm_leiden_res0.7'].replace({
    '0' : 'IN-RELN',
    '1' : 'IN-PVALB',
    '2' : 'IN-VIP',
    '3' : 'IN-RELN',
    '4' : 'IN-SV2C',
    '5' : 'IN-PVALB',
    '6' : 'IN-PVALB',
    '7' : 'IN-SV2C',
    '8' : 'IN-VIP',
    '9' : 'IN-GRIK1',
    '10' : 'IN-PVALB',
    '11' : 'IN-SST',
    '12' : 'IN-PVALB'  
})

mudata['rna'].obs['annotation_level_3'] = mudata['rna'].obs['mm_leiden_res0.7'].replace({
    '0' : 'IN-RELNpos_ROBO2hi_KIF26Bhi',
    '1' : 'IN-PVALB_MYO5Bhi_PLCL1hi',
    '2' : 'IN-VIPhi_GALNT17hi',
    '3' : 'IN-RELN_ADARB2hi_NCAM2hi',
    '4' : 'IN-SV2C_NRG1hi',
    '5' : 'IN-PVALB_TRPM3_KHDRBS2',
    '6' : 'IN-PVALB_HTR1Fhi',
    '7' : 'IN-SV2C_NOS1pos',
    '8' : 'IN-RELN_OLFM3hi',
    '9' : 'IN-THSD7Bhi_FBN2hi', # neurogenesis markers
    '10' : 'IN-PVALB_GRIK1pos_MYO5Blo_PLCL1lo',
    '11' : 'IN-SST_GULP1hi_IL1RAPL2_NOS1hi',
    '12' : 'IN-PVALB_C3'
    
})
```
