# 1.0 First checks

```python
import muon as mu
import numpy as np
import scanpy as sc
import matplotlib.pyplot as plt
import pandas as pd
```

```python
adata = mu.read('Neurons_Excitatory_clustered.h5mu/rna')
```

```python
import pandas as pd

mm_leiden_res07 = pd.read_csv('multimodal/algleiden_res0.7/clusters.txt.gz', sep = '\t')
mm_umap = pd.read_csv('multimodal/md0.5_umap.txt.gz', sep = '\t')
```

```python
adata.obsm['X_umap'] = np.array(mm_umap[['0', '1']])
```

```python
adata.obs['mm_leiden_res07'] = list(mm_leiden_res07['clusters'])

adata.obs['mm_leiden_res07'] = adata.obs['mm_leiden_res07'].astype('category')
```

```python
sc.pl.umap(adata, color = ['mm_leiden_res07','Group','CUX2', 'RORB','THEMIS', 'FEZF2', 'ID2'], s = 5)
```

```python
tmp = pd.crosstab(adata.obs['mm_leiden_res07'],adata.obs['sample_id'],normalize='index')
tmp.plot.bar(stacked=True).legend(bbox_to_anchor=(1.0, 1.0))

tmp = pd.crosstab(adata.obs['mm_leiden_res07'],adata.obs['Group'],normalize='index')
tmp.plot.bar(stacked=True).legend(bbox_to_anchor=(1.0, 1.0))
```
