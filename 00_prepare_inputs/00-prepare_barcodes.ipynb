```python
sample_list = glob.glob(path+"*")
sample_list.remove(path+'scripts_cellranger')
sample_list.remove(path+'data')
sample_list.remove(path+'multiome_pilot_aggr')
sample_list.remove(path+'C074C')
```

```python
for i in range (0,len(sample_list)):  
    sample=i+1
    if i==0:
        df = pd.read_csv(sample_list[i]+'/outs/per_barcode_metrics.csv')
        print(np.shape(df))
    else:
        df2 = pd.read_csv(sample_list[i]+'/outs/per_barcode_metrics.csv')
        df2['barcode'] = df2['barcode'].str.split('-', expand=True)[0]+'-%i'%sample
        df2['atac_barcode'] = df2['atac_barcode'].str.split('-', expand=True)[0]+'-%i'%sample
        df2['gex_barcode'] = df2['barcode']
        df=pd.concat([df,df2])
        
        print(np.shape(df))
print('final shape is ', np.shape(df))
```
